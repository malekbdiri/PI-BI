```python
# ✅ 1. Installation des dépendances si nécessaire
try:
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from statsmodels.tsa.arima.model import ARIMA
    from statsmodels.tsa.statespace.sarimax import SARIMAX
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.metrics import mean_squared_error
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dense
    print("✅ Toutes les dépendances sont déjà installées")
except ImportError as e:
    print(f"Installation des dépendances manquantes... {e}")
    import sys
    import subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", 
                         "pandas", "numpy", "matplotlib", 
                         "statsmodels", "scikit-learn", "tensorflow"])
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from statsmodels.tsa.arima.model import ARIMA
    from statsmodels.tsa.statespace.sarimax import SARIMAX
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.metrics import mean_squared_error
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dense

# ✅ 2. Connexion SQL & Chargement des données
def load_data():
    try:
        import pyodbc
        conn = pyodbc.connect(
            'DRIVER={ODBC Driver 17 for SQL Server};'
            'SERVER=DESKTOP-69VTGCD;'
            'DATABASE=DW_supplyChain;'
            'UID=sa;'
            'PWD=123456789123456789'
        )
        print("✅ Connexion SQL réussie")
        
        query = """
        SELECT manufacturedate, quantityused
        FROM dbo.Dim_Product
        WHERE manufacturedate IS NOT NULL AND quantityused IS NOT NULL
        """
        df = pd.read_sql(query, conn)
        
        # Nettoyage des données
        df['manufacturedate'] = pd.to_datetime(df['manufacturedate'], errors='coerce')
        df['quantityused'] = pd.to_numeric(df['quantityused'], errors='coerce')
        df.dropna(inplace=True)
        
        # Agrégation mensuelle
        ts = df.groupby(pd.Grouper(key='manufacturedate', freq='M'))['quantityused'].sum()
        ts.dropna(inplace=True)
        
        return ts
    
    except Exception as e:
        print(f"Erreur de chargement des données: {e}")
        # Chargement de données exemple si la connexion échoue
        print("Utilisation de données exemple...")
        dates = pd.date_range(start='2018-01-01', end='2023-12-31', freq='M')
        values = np.random.randint(1000, 5000, size=len(dates)) * np.linspace(1, 1.5, len(dates))
        ts = pd.Series(values, index=dates)
        return ts

ts = load_data()

# ✅ 3. Visualisation des données
plt.figure(figsize=(12, 6))
ts.plot(title="Série Temporelle - Quantity Used", ylabel="Quantity", xlabel="Date", grid=True)
plt.show()

# ✅ 4. Modèles ARIMA et SARIMA
def run_arima_models(ts):
    # ARIMA
    arima_model = ARIMA(ts, order=(1,1,1))
    arima_result = arima_model.fit()
    forecast_arima = arima_result.predict(start=ts.index[0], end=ts.index[-1], typ='levels')
    
    # SARIMA
    sarima_model = SARIMAX(ts, order=(1,1,1), seasonal_order=(1,1,1,12))
    sarima_result = sarima_model.fit()
    forecast_sarima = sarima_result.predict(start=ts.index[0], end=ts.index[-1])
    
    # Calcul des RMSE
    rmse_arima = np.sqrt(mean_squared_error(ts, forecast_arima))
    rmse_sarima = np.sqrt(mean_squared_error(ts, forecast_sarima))
    
    return forecast_arima, forecast_sarima, rmse_arima, rmse_sarima

forecast_arima, forecast_sarima, rmse_arima, rmse_sarima = run_arima_models(ts)

# ✅ 5. Modèle LSTM
def run_lstm_model(ts, time_step=12, epochs=100):
    # Normalisation
    scaler = MinMaxScaler(feature_range=(0, 1))
    ts_scaled = scaler.fit_transform(ts.values.reshape(-1, 1))
    
    # Création des séquences
    def create_dataset(data, time_step=1):
        X, y = [], []
        for i in range(len(data)-time_step-1):
            X.append(data[i:(i+time_step), 0])
            y.append(data[i + time_step, 0])
        return np.array(X), np.array(y)
    
    X, y = create_dataset(ts_scaled, time_step)
    X = X.reshape(X.shape[0], X.shape[1], 1)
    
    # Division train/test
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[0:train_size], X[train_size:len(X)]
    y_train, y_test = y[0:train_size], y[train_size:len(y)]
    
    # Construction du modèle
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
    model.add(LSTM(50))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    # Entraînement
    history = model.fit(X_train, y_train, 
                        validation_data=(X_test, y_test), 
                        epochs=epochs, 
                        batch_size=32, 
                        verbose=0)
    
    # Prédiction complète
    full_predict = model.predict(X)
    full_predict = scaler.inverse_transform(full_predict)
    lstm_forecast = pd.Series(full_predict[:,0], index=ts.index[time_step+1:len(ts)-1])
    
    # Calcul du RMSE
    common_index = ts.index.intersection(lstm_forecast.index)
    rmse_lstm = np.sqrt(mean_squared_error(ts.loc[common_index], lstm_forecast.loc[common_index]))
    
    return lstm_forecast, rmse_lstm, history

lstm_forecast, rmse_lstm, lstm_history = run_lstm_model(ts, epochs=50)

# ✅ 6. Visualisation des résultats
plt.figure(figsize=(14, 7))
ts.plot(label='Données Observées', color='black', linewidth=1)
forecast_arima.plot(label=f'ARIMA (RMSE={rmse_arima:.2f})', color='green', alpha=0.7)
forecast_sarima.plot(label=f'SARIMA (RMSE={rmse_sarima:.2f})', color='red', alpha=0.7)
lstm_forecast.plot(label=f'LSTM (RMSE={rmse_lstm:.2f})', color='blue', alpha=0.7)
plt.title('Comparaison des Modèles de Prévision')
plt.ylabel('Quantity Used')
plt.xlabel('Date')
plt.legend()
plt.grid(True)
plt.show()

# ✅ 7. Courbe d'apprentissage du LSTM
plt.figure(figsize=(10, 5))
plt.plot(lstm_history.history['loss'], label='Train Loss')
plt.plot(lstm_history.history['val_loss'], label='Validation Loss')
plt.title('Courbe d\'Apprentissage du Modèle LSTM')
plt.ylabel('Loss (MSE)')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
plt.show()

# ✅ 8. Résultats finaux
print("\n📊 Comparaison des Performances:")
print(f"ARIMA  - RMSE: {rmse_arima:.2f}")
print(f"SARIMA - RMSE: {rmse_sarima:.2f}")
print(f"LSTM   - RMSE: {rmse_lstm:.2f}")

# ✅ 9. Prévisions futures (optionnel)
def forecast_future(model, last_sequence, scaler, steps=12):
    future_predictions = []
    current_sequence = last_sequence.copy()
    
    for _ in range(steps):
        next_pred = model.predict(current_sequence.reshape(1, -1, 1))
        future_predictions.append(next_pred[0,0])
        current_sequence = np.append(current_sequence[1:], next_pred)
    
    return scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Exemple d'utilisation pour LSTM:
# scaler = MinMaxScaler().fit(ts.values.reshape(-1, 1))
# last_seq = scaler.transform(ts.values[-time_step:].reshape(-1, 1))
# future_pred = forecast_future(model, last_seq, scaler, steps=12)
```
